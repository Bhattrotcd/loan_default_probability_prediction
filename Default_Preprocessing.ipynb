```python
from datetime import datetime

temp_df['DATE_OF_BIRTH'] = pd.to_datetime(temp_df['DATE_OF_BIRTH'], format = '%d-%m-%Y', errors = 'coerce')
temp_df['DISBURSAL_DATE'] = pd.to_datetime(temp_df['DISBURSAL_DATE'], format = '%d-%m-%Y', errors = 'coerce')


def calculate_age(birth_date, disbursal_date):
    if pd.isnull(birth_date):
        return None  
    age = disbursal_date.year - birth_date.year - ((disbursal_date.month, disbursal_date.day) < (birth_date.month, birth_date.day))
    return age

temp_df['AGE'] =  temp_df.apply(lambda row: calculate_age(row['DATE_OF_BIRTH'], row['DISBURSAL_DATE']), axis=1)
```

```python
### Convert AVERAGE_ACCT_AGE and CREDIT_HISTORY_LENGTH to months
```

```python
def extract_months(age_str):
    if pd.isnull(age_str):
        return 0 
    years, months = 0, 0
    if 'yrs' in age_str:
        years = int(age_str.split('yrs')[0].strip())
    if 'mon' in age_str:
        months = int(age_str.split('yrs')[-1].split('mon')[0].strip())
    return (years * 12) + months

temp_df['AVERAGE_ACCT_AGE'] = temp_df['AVERAGE_ACCT_AGE'].apply(extract_months)
temp_df['CREDIT_HISTORY_LENGTH'] = temp_df['CREDIT_HISTORY_LENGTH'].apply(extract_months)
```

```python
import re
```

```python
def calculate_woe_iv(df,feature, target, bins = 10):
    if pd.api.types.is_numeric_dtype(df[feature]):
        df['bin'] = pd.qcut(df[feature], q = bins, duplicates = 'drop')
    else:
        df['bin'] = df[feature]

    woe_df = df.groupby('bin')[target].agg(['count','sum'])
    woe_df.columns = ['total', 'event']
    woe_df['non_event'] = woe_df['total'] - woe_df['event']

    woe_df['event_rate'] = woe_df['event'] / woe_df['event'].sum()
    woe_df['non_event_rate'] = woe_df['non_event'] / woe_df['non_event'].sum()

    woe_df['woe'] = np.log(woe_df['non_event_rate']/woe_df['event_rate']).replace([np.inf, -np.inf],0)
    woe_df['iv'] = (woe_df['non_event_rate'] - woe_df['event_rate'])*woe_df['woe']

    iv = woe_df['iv'].sum()

    df.drop(columns = 'bin', inplace = True)
    return iv

iv_values = {}
target_variable = 'LOAN_DEFAULT'
excluded_cols = ['UNIQUEID', target_variable]

for column in temp_df.columns:
    if column not in excluded_cols:
        iv = calculate_woe_iv(temp_df, column, target_variable, bins = 10)
        iv_values[column] = iv

sorted_iv = sorted(iv_values.items(), key = lambda x:x[1], reverse = True)
for variable, iv in sorted_iv:
    print(f"Variable: {variable}, Information Value: {iv}")
```

```python
####Quantile based binning does not give a good IV for any of the variable
####Trying to check the IV using binning technique based on decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
class MonotonicBinning:
    def __init__(self, max_leaf_nodes=5, min_samples_leaf=0.05):
        self.max_leaf_nodes = max_leaf_nodes
        self.min_samples_leaf = min_samples_leaf
        self.bins = None

    def fit(self, X, y):
        # Train decision tree to find bin boundaries
        tree = DecisionTreeClassifier(max_leaf_nodes=self.max_leaf_nodes, min_samples_leaf=self.min_samples_leaf)
        tree.fit(X.values.reshape(-1, 1), y)
        thresholds = tree.tree_.threshold[tree.tree_.threshold != -2]
        self.bins = np.sort(thresholds)
        self.bins = np.concatenate(([X.min() - 1], self.bins, [X.max() + 1]))
        return self

    def transform(self, X):
        # Digitize data based on bin edges
        return np.digitize(X, self.bins)

    def get_bin_edges(self):
        return self.bins

# Function to calculate WoE and IV for a binned variable
def calculate_woe_iv(binned_data, target):
    woe_dict = {}
    iv = 0
    for bin_id in np.unique(binned_data):
        in_bin = (binned_data == bin_id)
        event_rate = target[in_bin].mean()
        non_event_rate = 1 - event_rate

        # Calculate WoE
        if event_rate == 0 or non_event_rate == 0:
            woe = 0  # Handle no events or non-events in the bin
        else:
            woe = np.log(non_event_rate / event_rate)

        # Store WoE and calculate IV
        woe_dict[bin_id] = woe
        iv += (non_event_rate - event_rate) * woe
    return iv

# List of continuous variables to bin
continuous_vars_to_bin = [
    'DISBURSED_AMOUNT', 'ASSET_COST', 'LTV', 'PERFORM_CNS_SCORE', 'PRI_NO_OF_ACCTS', 'PRI_ACTIVE_ACCTS', 
    'PRI_OVERDUE_ACCTS', 'PRI_CURRENT_BALANCE', 'PRI_SANCTIONED_AMOUNT', 'PRI_DISBURSED_AMOUNT', 
    'SEC_NO_OF_ACCTS', 'SEC_ACTIVE_ACCTS', 'SEC_OVERDUE_ACCTS', 'SEC_CURRENT_BALANCE', 
    'SEC_SANCTIONED_AMOUNT', 'SEC_DISBURSED_AMOUNT', 'PRIMARY_INSTAL_AMT', 'SEC_INSTAL_AMT', 
    'NEW_ACCTS_IN_LAST_SIX_MONTHS', 'DELINQUENT_ACCTS_IN_LAST_SIX_MONTHS', 'AVERAGE_ACCT_AGE', 
    'CREDIT_HISTORY_LENGTH', 'AGE', 'NO_OF_INQUIRIES'
]

# Target variable
target_variable = 'LOAN_DEFAULT'

# Dictionary to store IV values for each variable
iv_summary = {}

# Loop through each variable in the dataset
for var in temp_df.columns:
    # Check if the variable needs binning
    if var in continuous_vars_to_bin:
        # Apply monotonic binning using decision tree
        try:
            monotonic_binning = MonotonicBinning(max_leaf_nodes=5)
            monotonic_binning.fit(temp_df[var], temp_df[target_variable])
            binned_data = monotonic_binning.transform(temp_df[var])
            iv_value = calculate_woe_iv(binned_data, temp_df[target_variable])
        except Exception as e:
            print(f"Error processing {var}: {e}")
            iv_value = np.nan
    elif var not in [target_variable, 'UNIQUEID']:
        # For non-binned variables, treat them as categorical and calculate WoE directly
        try:
            binned_data = pd.factorize(temp_df[var])[0]  # Factorize categorical variables
            iv_value = calculate_woe_iv(binned_data, temp_df[target_variable])
        except Exception as e:
            print(f"Error processing {var}: {e}")
            iv_value = np.nan
    else:
        iv_value = np.nan  # Skip target and ID variables

    # Store the IV for the variable
    iv_summary[var] = iv_value

# Convert IV summary to DataFrame
iv_df = pd.DataFrame(list(iv_summary.items()), columns=['Variable', 'IV']).sort_values(by='IV', ascending=False)

# Display the IV summary DataFrame
iv_df
```

```python
# Function to map PIN code to region/state based on the first two digits
def map_state_region(pincode):
    # Ensure pincode is valid and get the first two digits as an integer
    if pd.notnull(pincode):
        first_two_digits = int(str(pincode)[:2])
        
        # Map based on first two digits range
        if first_two_digits == 11:
            return 'Delhi'
        elif first_two_digits in [12, 13]:
            return 'Haryana'
        elif first_two_digits in [14, 15, 16]:
            return 'Punjab'
        elif first_two_digits == 17:
            return 'Himachal Pradesh'
        elif first_two_digits in [18, 19]:
            return 'Jammu & Kashmir'
        elif 20 <= first_two_digits <= 28:
            return 'Uttar Pradesh and Uttaranchal'
        elif 30 <= first_two_digits <= 34:
            return 'Rajasthan'
        elif 36 <= first_two_digits <= 39:
            return 'Gujarat'
        elif 40 <= first_two_digits <= 44:
            return 'Maharashtra'
        elif 45 <= first_two_digits <= 49:
            return 'Madhya Pradesh and Chattisgarh'
        elif 50 <= first_two_digits <= 53:
            return 'Andhra Pradesh & Telangana'
        elif 56 <= first_two_digits <= 59:
            return 'Karnataka'
        elif 60 <= first_two_digits <= 64:
            return 'Tamil Nadu'
        elif 67 <= first_two_digits <= 69:
            return 'Kerala'
        elif 70 <= first_two_digits <= 74:
            return 'West Bengal'
        elif first_two_digits in [75, 76, 77]:
            return 'Orissa'
        elif first_two_digits == 78:
            return 'Assam'
        elif first_two_digits == 79:
            return 'North Eastern'
        elif 80 <= first_two_digits <= 85:
            return 'Bihar and Jharkhand'
        elif 90 <= first_two_digits <= 99:
            return 'Army Postal Service (APS)'
        else:
            return 'Unknown'  # Default if no match is found
    else:
        return 'Unknown'  # For NaN or invalid pincode values

# Apply the state/region mapping to the 'CURRENT_PINCODE_ID' column
iv_df1['Region'] = iv_df1['CURRENT_PINCODE_ID'].apply(map_state_region)

# Display the DataFrame with the new 'Region' column
print(iv_df1[['CURRENT_PINCODE_ID','Region']])
```
